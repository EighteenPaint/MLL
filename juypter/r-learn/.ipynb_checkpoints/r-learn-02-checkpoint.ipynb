```r
learn = sleep  #使用sleep数据集进行学习
```

```r
is.data.frame(learn) #判断数据是否是dataframe类型
```

```r
mat = matrix(1:32,4,8,byrow = T);#创建一个4 X 8的矩阵
mat 
```

```r
is.data.frame(mat)  #这里肯定会变为false
```

```r
mat2frame = as.data.frame(mat);# 将matrix转化为dateframe
mat2frame  
```

<strong style="color:red">注意：frame转化为matrix较为复杂，因为frmae类型会存在不一致的元素</strong>

```r
sample(mat2frame$V1,3,replace=T)  #有放回抽取
sample(mat2frame$V1,3,replace=F)  #无放回抽取
```

```r
mat2frame
subset(mat2frame,V1 > 40 , select=V2)  # V1 > 40
subset(mat2frame,V1 > 10 , select=V2)  #V1 > 10
```

```r
sd(mat2frame$V1)
```
