```python
#从文件读取的数据要记得转成series
org_layer_df = pd.read_csv("org_to_ids_new_list.csv")
org_layer_df.rename(columns={'Unnamed: 0':'org_name','0':'article_id'},inplace=True)
# org_layer_df['article_id'].head(10)
org_layer_series = pd.Series(data=org_layer_df['article_id'].values,index=org_layer_df['org_name'])
org_layer_series.head(10)
```

```python
#{org_name:{name1:[ids]}}
org_to_name = dict()
for org_name in org_layer_series.index:  # 遍历每一个机构簇
    same_name_cluster = dict() # 对于每一个机构需要生成机构下的同名簇
    for article_id in org_layer_series[org_name].split(","): #每一个机构簇里所有论文
        authors = data[article_id]['authors'] # 取出文章的作者数据
        for name_dict in authors: #遍历该篇文章的作者数据
            author_name = name_dict['name'] #取出name字段
            if author_name not in same_name_cluster.keys(): # 检查是不是第一次
                same_name_cluster[author_name] = set()  #不要用set直接放入字符串，会被变成字符
                same_name_cluster[author_name].add(article_id)
            else:#如果已经存在，则直接增加即可
                same_name_cluster[author_name].add(article_id)
    org_to_name[org_name] = same_name_cluster # 挂载在机构下，形成机构下同名簇
```

```python
otn_df = pd.DataFrame(org_to_name.items())
otn_df.head(10) #转化成dataframe，到这个环节你会发现，对内存和计算的需求都在上升，为了变小其实可以直接使用values，不要机构名称
```

```python
def cluster(result,name,article_ids,article_keywords_map):
    keywords = []
    for article_id in article_ids:
        keywords.append(article_keywords_map[article_id])
     #训练关键词语料库
    with open("corpus") as file:  # 使用保存的语料库，不用反复读取文件操作，提高效率
        corpus = file.read()
    from sklearn.feature_extraction.text import CountVectorizer
    vecizer = CountVectorizer(stop_words='english', token_pattern=r"(?u)\b\S+\b")  # 由于专业名词的原因，最好不要使用默认的token pattern
    vecizer.fit([corpus])  # 进行训练
    #将关键词向量化
    key_word_vector=[]
    for keyword in keywords[0:100]:
        wordFlatten = [" ".join(keyword)]
        vetctor = vecizer.transform(wordFlatten).toarray()[0]
        key_word_vector.append(vetctor)
    #进行聚类
    from sklearn.cluster import DBSCAN
    y_pred = DBSCAN(eps=5, min_samples=3).fit_predict(key_word_vector) #todo: eps和minsample没有确定
    #todo:生成结果
    pred_result = dict()
    for i in range(0, len(y_pred)):
        label = y_pred[i]
        if label not in pred_result.keys():
            pred_result[label] = []
            pred_result[label].extend(article_ids[i])
        else:
            pred_result[label].extend(article_ids[i])
    for value in pred_result.values():
        result[name].append(value)
```

```python
# result:{author:[[ids1][ids2][ids3]......]}
#todo:生成article_keywords_map
article_keywords_map = dict()
for column in data.columns:
    keyword = data[column]['keywords']
    if isinstance(keyword,list):
        article_keywords_map[column] = keyword
    else:
        article_keywords_map[column] = " "
result = dict()
for names_to_ids_dict in  otn_sample_form_json['1']: #取出第二列
    for name in names_to_ids_dict.keys(): # 遍历字典
        ids_for_each_name = names_to_ids_dict[name]
        if(len(ids_for_each_name) == 1):
            if name not in result.keys(): #如果没有则新建，并append
                result[name] = []
                result[name].append(ids_for_each_name)
            else:
                result[name].append(ids_for_each_name)
        else:#如果有多个，则需要进行聚类算法操作
            cluster(result,name,ids_for_each_name,article_keywords_map)
```
